```python
#실습 1
import matplotlib.pyplot as plt
import csv
f=open('seoul10.csv')
data=csv.reader(f)
next(data)
max=[]
min=[]
for row in data:
    if row[-1]!='':
        if row[0].split('/')[1]=='10' and row[0].split('/')[2]=='11':
            max.append(float(row[-1]))

f.seek(0)
data= csv.reader(f)
header=next(data)
            
for row in data:
    if row[-2]!='':
        if row[0].split('/')[1]=='10' and row[0].split('/')[2]=='11':
            min.append(float(row[-2]))
            
plt.plot(max,'pink')
plt.plot(min,'skyblue')
plt.show()
```

```python
#실습2
import matplotlib.pyplot as plt
import csv
f=open('seoul10.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('/')[1]
    if row[-1] !='':
        if month=='01':
            jan.append(float(row[-1]))
plt.hist(jan,bins=100,color='b',label='Jan')

f.seek(0)
data= csv.reader(f)
header=next(data)
            
for row in data:
    month=row[0].split('/')[1]
    if row[-1] !='':
        if month=='08':
            aug.append(float(row[-1]))
            
plt.hist(aug,bins=100,color='r',label='Aug')
plt.legend()
plt.show()
```
