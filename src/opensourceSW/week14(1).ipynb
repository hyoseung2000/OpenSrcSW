```python
#week14
import pandas as pd
import math
import matplotlib.pyplot as plt

df=pd.read_csv('population.csv',header=3)

gwangjin=0
dd=df['행정구역'].str.split(' ')
for i in range(0,len(dd),1):
    if(dd[i][1]=='광진구'):
        gwangjin=i
        
gjinpop=[]
for m in range(101):
    gjinpop.append(int(df.loc[gwangjin][3+m].replace(',','')))
    
gap=[]

#행읽어오기
for j in range(1,len(dd),1):
    total=0
    #0-100세
    for k in range(101):
        p1=int(df.loc[j][3+k].replace(',',''))
        total+=((p1-gjinpop[k])*(p1-gjinpop[k]))
    gap.append(total)

#가장 큰 차이
max=-1
maxnum=0
maxpop=[]
for g in range (len(dd)-1):
    if(gap[g]>max):
        max=gap[g]
        maxnum=g
        
for a in range(101):
    maxpop.append(int(df.loc[maxnum+1][3+a].replace(',','')))
        
#가장 작은 차이
min=999999999999
minnum=0
minpop=[]
for h in range (len(dd)-1):
    if(gap[h]<min):
        if(gap[h]==0):
            continue
        min=gap[h]
        minnum=h
            
for b in range(101):
    minpop.append(int(df.loc[minnum+1][3+b].replace(',','')))

plt.rc('font',family='Malgun Gothic')
plt.style.use('ggplot')
plt.plot(gjinpop,color='black',label='광진구')
plt.plot(maxpop,color='red',label=(str(dd[maxnum+1][1])+'( '+str(math.sqrt(float(max)))+' )'))
plt.plot(minpop,color='blue',label=(str(dd[minnum+1][1])+'( '+str(math.sqrt(float(min)))+' )'))
plt.legend()
plt.show()
        
        
            
        
    
        
    
```
